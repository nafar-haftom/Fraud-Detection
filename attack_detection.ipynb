```python
dataset_path = '/content/IoT-SDN IDS.csv'  # Adjust the path according to where you've stored the dataset
df = pd.read_csv(dataset_path)
print("First 5 Rows of the Dataset:")
df.head()
```

```python
print("\nDataset Summary (Statistical):")
print(df.describe())
```

```python
print("\nDataset Information (Structure):")
print(df.info())
```

```python
X = df.iloc[:, :-1]  # Features
y = df.iloc[:, -1]   # Labels
```

```python
scaler = MinMaxScaler(feature_range=(0, 100))

# Assuming X is your features DataFrame from earlier steps
# Replace inf/-inf with NaN and then fill in NaN values if you haven't already
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.fillna(X.mean(), inplace=True)  # Impute NaNs with mean, or choose another strategy

# Fit the scaler to your data and then transform it
X_scaled = scaler.fit_transform(X)
```

```python

# Applying PCA
pca = PCA(n_components=15)  # Adjust the number of components as needed
X_pca = pca.fit_transform(X_scaled)

# Splitting dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
# Models initialization = Decision Tree Classifier
dtc = DecisionTreeClassifier()
# Models training
dtc.fit(X_train, y_train)
# Making predictions
dtc_pred = dtc.predict(X_test)
# Evaluating the models
dtc_accuracy = accuracy_score(y_test, dtc_pred)
print(f'Decision Tree Classifier Accuracy: {dtc_accuracy}')
```

```python
# Models initialization = Random Forest Classifier
rfc = RandomForestClassifier()
# Models training
rfc.fit(X_train, y_train)
# Making predictions
rfc_pred = rfc.predict(X_test)
# Evaluating the models
rfc_accuracy = accuracy_score(y_test, rfc_pred)
print(f'Random Forest Classifier Accuracy: {rfc_accuracy}')
```

```python
# Models initialization = K-Nearest Neighbors
knn = KNeighborsClassifier()
# Models training
knn.fit(X_train, y_train)
# Making predictions
knn_pred = knn.predict(X_test)
# Evaluating the models
knn_accuracy = accuracy_score(y_test, knn_pred)
print(f'K-Nearest Neighbors Accuracy: {knn_accuracy}')
```

```python
# Models initialization = Logistic Regression
lgr = LogisticRegression()
# Models training
lgr.fit(X_train, y_train)
# Making predictions
lgr_pred = lgr.predict(X_test)
# Evaluating the models
lgr_accuracy = accuracy_score(y_test, lgr_pred)
print(f'Logistic Regression Accuracy: {lgr_accuracy}')
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, classification_report
import numpy as np

def calculate_metrics_multi_class(y_true, y_pred):
    # Precision, Recall for multi-class
    precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_true, y_pred, average='macro')

    # Detailed classification report (includes F1-score, support, and more for each class)
    class_report = classification_report(y_true, y_pred, zero_division=0)

    return {
        'Precision (Macro-Averaged)': precision,
        'Recall (Macro-Averaged)': recall,
        'Classification Report': class_report
    }

# Assuming y_test, dtc_pred, rfc_pred, knn_pred, and lgr_pred are defined
predictions = {
    'DTC': dtc_pred,
    'RFC': rfc_pred,
    'KNN': knn_pred,
    'LGR': lgr_pred
}

# Calculate and print metrics for each model
for model_name, model_pred in predictions.items():
    metrics = calculate_metrics_multi_class(y_test, model_pred)
    print(f'Metrics for {model_name}:')
    for metric_name, metric_value in metrics.items():
        if metric_name != 'Classification Report':
            print(f'{metric_name}: {metric_value}')
    # Print classification report separately for better formatting
    print(f'\nClassification Report for {model_name}:\n{metrics["Classification Report"]}\n')
```
